```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

from data import retrieve_news_cat_data, prepare_feature_data, vectorize, plot_category_dist
from classify import  perform_centroid_test, perform_knn_test, perform_svm_test
```

```python
news_data_frame, category_list = retrieve_news_cat_data()
```

```python
# uncomment to see scrubbed category distribution
plot_category_dist(news_data_frame)
```

```python
feature_data_frame = prepare_feature_data(news_data_frame, category_list)
```

```python
def plot_confusion_matrix(y_actual, y_pred,labels):
    cm = confusion_matrix(y_actual, y_pred,labels = labels)
    plt.figure(figsize=(10, 10))  # Sets the figure size
    sns.heatmap(cm, annot=True, fmt="d",xticklabels = labels,yticklabels = labels,)
    plt.ylabel('actual category')
    plt.xlabel('predicted category')
    plt.title('Confusion Matrix') 
    plt.show()  # Displays the plot
```

```python
X_train, X_test,y_train,y_test = vectorize(feature_data_frame,'news','category')
```

```python
centroid_pred, centroid_classes = perform_centroid_test(X_train, X_test,y_train,y_test)
plot_confusion_matrix(y_test, centroid_pred, centroid_classes)
```

```python
knn_pred, knn_classes = perform_knn_test(X_train, X_test,y_train,y_test)
plot_confusion_matrix(y_test, knn_pred, knn_classes)
```

```python
# NOTE: please be aware, this cell takes ~14hrs to complete on an M2 Mac, 32G
svm_pred, svm_classes = perform_svm_test(X_train, X_test,y_train,y_test)
plot_confusion_matrix(y_test, svm_pred, svm_classes)
```
